```python
import numpy as np
import matplotlib
import cv2
from matplotlib import pyplot as plt
```

```python
image  = cv2.imread('./Testing_Images/Serena.jpg')
image.shape
```

```python
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
gray
```

```python
plt.imshow(gray, cmap = 'gray')
```

```python
facecascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
faces = facecascade.detectMultiScale(gray,1.3,5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
face_img = cv2.rectangle(image,(x,y) , (x+w, y+h) , (255 , 0, 0), 2)
plt.imshow(face_img)
```

```python
eyecascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
    ri_gray = gray[y:y+h, x:x+w]
    ri_color = face_img[y:y+h, x:x+w]
    eyes = eyecascade.detectMultiScale(ri_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(ri_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
        

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```
